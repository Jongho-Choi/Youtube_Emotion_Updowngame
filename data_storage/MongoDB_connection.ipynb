### 4. MongoDB 데이터 불러오기

```python
from pymongo import MongoClient
client = MongoClient(host='3.37.202.252', port=27017)
DATABASE_NAME = 'youtube_comment_database'
COLLECTION_NAME = 'youtube_comment_collection'
database = client[DATABASE_NAME]
collection = database[COLLECTION_NAME]
```

```python
## 지정한 collection안에 데이터 전부 불러오기
for d in collection.find():
  print(d)
```

```python
## 가져올 id 확인
post_id_as_str = str('62bbb2d451ba3d33259a01bb')
post_id_as_str
```

```python
## id 그대로 입력하면 None이 뜨므로 아래와 같이 bson처리후 불러옴
from bson.objectid import ObjectId

bson_id = ObjectId(post_id_as_str)
video_comments = collection.find_one({"_id": bson_id})
```

```python
print(video_comments)
```
